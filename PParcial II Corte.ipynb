## 7.3.3 Deposición de superficie

Varios procesos naturales y artificiales implican el depósito de partículas en una superficie para formar una película. Debido a que las partículas se evaporan típicamente de un filamento caliente, hay aleatoriedad en el proceso de emisión; sin embargo, las películas producidas resultan tener estructuras bien definidas.

**1.** Cree una simulación Monte Carlo de partículas que caen y se adhieren a una línea horizontal de longitud L compuesta de 200 sitios de deposición.
* a. Comience todas las partículas desde la misma altura.
* b. Simule las diferentes velocidades de emisión haciendo que las partículas caigan a distancias aleatorias a lo largo de la línea.
* c. Haga que una partícula se adhiera al sitio donde aterriza.
* d. Debido a que habrá columnas de partículas depuestas de varias alturas, la trayectoria de una partícula puede bloquearse antes de llegar a la línea, o puede rebotar entre columnas y luego caer en un agujero. Por lo tanto, suponga que si la altura de la columna a la que cae la partícula es mayor que la de sus dos vecinos, se agregará a esa altura.
* e. Si la partícula cae en un hoyo, o si hay un hoyo adyacente, llena el hoyo.
* f. Realice un diagrama continuo a medida que se deposita la superficie.
